```python
# Step 1: Import the required libraries
import spacy
import networkx as nx
import matplotlib.pyplot as plt

# Step 2: Define the find_entities function
def find_entities(paragraph):
    # Load the Spacy English model
    nlp = spacy.load("en_core_web_sm")
    # Process the paragraph
    doc = nlp(paragraph)
    entities = []
    for entity in doc.ents:
        entities.append((entity.text, entity.label_))
    return entities

# Step 3: Define the find_relations function
def find_relations(entities):
    relations = []
    for i in range(len(entities)):
        for j in range(i + 1, len(entities)):
            relations.append((entities[i][0], entities[j][1], entities[j][0]))
    return relations

# Step 4: Define the display_knowledge_graph function
def display_knowledge_graph(relations):
    # Create an empty graph
    graph = nx.Graph()
    for relation in relations:
        subject, predicate, obj = relation
        graph.add_edge(subject, obj, label=predicate)

    # Generate the layout for the graph
    pos = nx.spring_layout(graph, k=0.2)
    labels = nx.get_edge_attributes(graph, 'label')

    # Draw the nodes, edges, and labels of the graph
    nx.draw_networkx(graph, pos, with_labels=True, node_size=2000, font_size=12, node_color='g')
    nx.draw_networkx_edge_labels(graph, pos, edge_labels=labels)

    # Hide the axis and display the graph
    plt.axis('off')
    plt.show()

# Example usage
paragraph = "The United States of America, commonly referred to as the USA or simply America, is a vast country located in North America. It is bordered by Canada to the north and Mexico to the south. With a rich history and diverse population, the United States has become a global powerhouse in various fields. Washington, D.C. is the capital city, while New York City is widely recognized as the financial and cultural hub of the nation. From the iconic Statue of Liberty to the breathtaking landscapes of the Grand Canyon and the bustling streets of cities like Los Angeles and Chicago, the United States offers a tapestry of natural wonders, vibrant cities, and a melting pot of cultures, making it a truly remarkable and influential nation on the global stage."
entities = find_entities(paragraph)
print("Entities:")
for entity in entities:
    print(entity)
print()

relations = find_relations(entities)
print("Relations:")
for relation in relations:
    print(relation)
print()

print("Knowledge Graph:")
display_knowledge_graph(relations)
```
